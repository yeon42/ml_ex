# Groupby 1

```python
import pandas as pd
```

## groupby - basic

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby("Team")["Points"].sum()
```

## Hierarchical index

```python
h_index = df.groupby(["Team", "Year"])["Points"].sum()
h_index
```

```python
h_index.index
```

```python
h_index["Devils": "Kings"]
```

```python
h_index.unstack()

# unstack() : group으로 묶여진 데이터를 matrix 형태로 전환해줌
```

```python
h_index.swaplevel()

# Year과 Team 바꾹;
```

```python
h_index.sum(level=0)
# index별로 더하기
```

## filter
- 특정 조건으로 데이터를 검색할 때 사용

```python
df.groupby('Team').filter(lambda x: len(x) >= 3) # 데이터가 3개 이상일 때

# len(x): grouped된 dataframe의 수
```

```python
df.groupby('Team').filter(lambda x: x["Points"].max() > 800)
```

# Case Study

```python
phone = {'index': [0, 1, 2, 3, 4],
        'date': ['15/10/14 06:58', '15/10/14 06:58', '15/10/14 14:46', '15/10/14 14:48', '15/10/14 17:27'],
        'duration': [34.429, 13.000, 23.000, 4.000, 4.000],
        'item': ['data', 'call', 'call', 'call', 'call'],
        'month': ['2014-11', '2014-11', '2014-11', '2014-11', '2014-11'],
        'network': ['data', 'Vodafone', 'Meteor', 'Tesco', 'Tesco'],
        'nework_type': ['data', 'mobile', 'mobile', 'mobile', 'mobile']
        }

df_phone = pd.DataFrame(phone)
df_phone
```

## Data

```python
import dateutil

df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
df_phone.groupby('month')['duration'].sum()
```

```python
df_phone[df_phone['item']=='call'].groupby('month')['duration'].sum()
```

```python
df_phone.groupby(['month', 'item'])['duration'].sum()
```

```python
df_phone.groupby(['month', 'item'])['date'].count().unstack()
```

```python
df_phone.groupby('month', as_index=False).agg({"duration": "sum"})
```
